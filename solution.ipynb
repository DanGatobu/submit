```python
def r_yes():
    if random.random() < 0.9:
        return "yes"
    else:
        return "no"
    
def update_value(current_value,allowed_ppl,decrease_rate):
    # Check if the current value is 0
    if current_value == 0:
        # There's a 75% chance it stays at 0
        if random.random() < allowed_ppl :
            return 'yes'
        else:
            return 'No'
    else:
        # Decrease the current value by 20%
        return current_value - (current_value * decrease_rate)
```

```python
def days_open():
    # Generate a random number between 1 and 1000
    num = random.randint(1, 1000)
    
    # Bias the random number towards higher values
    if random.random() < 0.85:
        num = random.randint(750, 1000)  # Adjust the range for higher values
        
    return num
```
