```python
#assisting function
def dc():
    if random.random() < 0.95:
        return "yes"
    else:
        return "no"
```

# ELIGIBLITY CALCULATOR FUNCTION
### function generates if a loan is legible depending on number of loans taken(either upsell or upgrades)
### aims to check if increase/decrese  from 75% allowed results to hihger results
### Also amims to chak if increase /decrease of the percentage decrese results in higher profit (initial set is 20%)h h
### only for cash loans


```python
def eligibity_calculator(num_loans,initial_allowed,d_rate):
    probability = initial_allowed

    
    probability -= num_loans * d_rate

    
    probability = max(0, min(1, probability))

   
    random_number = random.random()

    
    if random_number <= probability:
        return "yes"
    else:
        return "no"
```

# DAYS OPEN FUNCTION
### Function generates the number of days a is allowed to run
### all loans that are eligible have 400 days
### funtction assumes 85% of the users taking loans will be below the set limit 
### aims to beable to check if higher loan periods result to higher values


```python

def days_open(loan_period):
    loan_period=int(loan_period)
    
    if random.random() > 0.85:
        num = random.randint(loan_period, 1000)
    else:
        num=random.randint(0, loan_period)
        
    return num
```

#### Asumptions the loans initial principle is a value between 100 - 1000
#### Monthly sales return per user is calculated as firstdayprofit + monthly intrest rate
#### Monthly intrest rate= yearly intest rate/12
#### Function returns monthly expected profit(volume)for a user under specific conditions based on :
* initial_loan status (can be either finished payment or unfinished)
* initial_loan_type(if the initial laon is upgrade and then upsell loan type is not possible )
* loan_period (if loan closed not eligible)
#### Asumption for  loan running time calculation when a person is applying for either an upgrade or upsell 85% of time they will be have ther required limit
* up_time  (variable to be rerurned to later check if shorter loan perods will result to higher profits)
* loan_type (can either be cash for cash loans upgrade for any loan that is of type upgrade (product upgrades))
* cash_loan_volume_limit (variable to be user to check later if putting a limit on cash loans will result in higher profit overall)
* upgrade_loan_volume_limit (variable to be user to check later if putting a limit on cash loans will result in higher profit overall)
* existing num_loans (sets the existing loans for eligiblity calculator function)
* existing num loans (sets how many loans does the person currently have (for  upsell and  upgrade) more== less chances)
* allowed_ppl (number of people who are allowed to take loan)
* decrease rate( rate at which  probablility of loan being acceped reduces per number of taken loans )
* accept upgrade (default parameter of 0.5 to reprecent what percent of customers accept an upgrade)
* accept cash (default parameter of 0.9 to reprecent what percent of customers accept an cash upgrade)

* lv (loan volume Directly propotional to profit so default will always be 1 in testing )





```python


def loan_calculator(initial_loan_status,initial_loan_type,loan_period, loan_type, cash_loan_volume_limit, upgrade_loan_volume_limit, existing_num_loans, allowed_ppl, decrease_rate,accept_cash=0.9,accept_upgrade=0.9 ,cash_loan_acceptance=0.9,upgrade_loan_acceptance=0.5,lv=1):
    loan_running_time = days_open(loan_period)# calculate days open assumption is over 85% the duration is always okay
    eligible = True
    loan_volume = lv
    take = True
    amount = 0
    lm_amount = random.randint(100,1000)
    rm_amount = 0
    rt_lst = []
    
    if loan_running_time > loan_period:
        eligible = False
        
#     initial_loan_time = finished_date + 50
    
    if initial_loan_type == 'upgrade' and loan_type=='cash':
        eligible = False
        
    if initial_loan_status == 'finished':
        eligible = False
        
     # not sure what this variable is supposed to be
    
    if loan_type == 'upgrade':
        if dc()=='no':# replaced close dates to always be 95% yes since no dates
            eligible = False
    
    if loan_type == 'cash':
        if random.random() > cash_loan_acceptance:
            eligible = False
            
        
        if loan_volume > cash_loan_volume_limit:
            eligible = False
    
    if loan_type == 'upgrade':
        if random.random() > upgrade_loan_acceptance:
            eligible = False
        if loan_volume > upgrade_loan_volume_limit:
            eligible = False
     #to be changed later to check how 
        
    if loan_type=='cash':

        if eligibity_calculator(existing_num_loans, allowed_ppl, decrease_rate) == 'No':
            eligible = False
    
    if loan_type == 'cash':
        if random.random() > accept_cash:
            take = False
            
    if loan_type == 'upgrade':
        if random.random() > accept_upgrade:
            take = False
    if loan_type == 'cash' and eligible and take:
        amount = 25 * loan_volume
        amount_monthly = lm_amount * 0.0125
        rm_amount = amount + amount_monthly
        rt_lst.append((loan_type, rm_amount))
      
    if loan_type == 'upgrade' and eligible and take:
        amount = 150 * loan_volume
        amount_monthly = lm_amount * 0.0125
        rm_amount = amount + amount_monthly
        rt_lst.append((loan_type, rm_amount))
        
    
    return rt_lst, rm_amount
```

```python


def loan_calculator_optimizer(initial_loan_status,initial_loan_type,loan_period, loan_type, cash_loan_volume_limit, upgrade_loan_volume_limit, existing_num_loans, allowed_ppl, decrease_rate,accept_cash=0.9,accept_upgrade=0.9 ,cash_loan_acceptance=0.9,upgrade_loan_acceptance=0.5,lv=1):
    loan_running_time = days_open(loan_period)# calculate days open assumption is over 85% the duration is always okay
    eligible = True
    loan_volume = lv
    take = True
    amount = 0
    lm_amount = random.randint(100,1000)
    rm_amount = 0
    rt_lst = []
    
    if initial_loan_status==1:
        initial_loan_status='finished'
    else:
        initial_loan_status='unfinished'
        
    if loan_type==1:
        loan_type='cash'
    else:
        loan_type='upgrade'
    
    if loan_running_time > loan_period:
        eligible = False
        
    
    
    if initial_loan_type == 'upgrade' and loan_type=='cash':
        eligible = False
        
    if initial_loan_status == 'finished':
        eligible = False
        
     # not sure what this variable is supposed to be
    
    if loan_type == 'upgrade':
        if dc()=='no':# replaced close dates to always be 95% yes since no dates
            eligible = False
    
    if loan_type == 'cash':
        if random.random() > cash_loan_acceptance:
            eligible = False
            
        
        if loan_volume > cash_loan_volume_limit:
            eligible = False
    
    if loan_type == 'upgrade':
        if random.random() > upgrade_loan_acceptance:
            eligible = False
        if loan_volume > upgrade_loan_volume_limit:
            eligible = False
     #to be changed later to check how 
        
    if loan_type=='cash':

        if eligibity_calculator(existing_num_loans, allowed_ppl, decrease_rate) == 'No':
            eligible = False
    
    if loan_type == 'cash':
        if random.random() > accept_cash:
            take = False
            
    if loan_type == 'upgrade':
        if random.random() > accept_upgrade:
            take = False
    if loan_type == 'cash' and eligible and take:
        amount = 25 * loan_volume
        amount_monthly = lm_amount * 0.0125
        rm_amount = amount + amount_monthly
        rt_lst.append((loan_type, rm_amount))
      
    if loan_type == 'upgrade' and eligible and take:
        amount = 150 * loan_volume
        amount_monthly = lm_amount * 0.0125
        rm_amount = amount + amount_monthly
        rt_lst.append((loan_type, rm_amount))
        
    
    return -rm_amount
```

```python
import time
from skopt import gp_minimize
from skopt.space import Real
from skopt.space import Categorical
```

### Using gp_mimimize to find optimal parameters
### Setting up of variables to use when optimizing the calcutator function
### Assumptions here cah and upgrade volumes are nevr above 10 since ther are directly propotional to volume

```python
pram_info=[]
```

```python
param_space = [
    
    Categorical([1, 0], name='loan_status'),# 1 for finished 0 for not finished
    
    Categorical([1, 0], name='in_loantype'),# 1 for cash 0 for upgrade
    
    Real(1,1000, name='loanperiod'),
    
    Categorical([1, 0], name='loan_type'),## 1 for cash 0 for upgarade
    
    Real(1,10, name='cash_loan_volume'),
    Real(1,10, name='upgrade_loan_volume'),
    Real(1,10, name='existing_num_loans'),
    Real(0.1,0.95, name='allowed_ppl'),
    Real(0.1,0.65, name='decrease_rate'),
    
]


pram_info=[]

def ob_lv(params):
    loan_status, in_loantype, loanperiod, loan_type, cash_loan_volume, upgrade_loan_volume, existing_num_loans, allowed_ppl, decrease_rate= params
    print("Parameters:")
    print("Loan Status:", loan_status)
    print("In Loan Type:", in_loantype)
    print("Loan Period:", loanperiod)
    print("Loan Type:", loan_type)
    print("Cash Loan Volume:", cash_loan_volume)
    print("Upgrade Loan Volume:", upgrade_loan_volume)
    print("Existing Number of Loans:", existing_num_loans)
    print("Allowed People:", allowed_ppl)
    print("Decrease Rate:", decrease_rate)
  
    result = loan_calculator_optimizer(loan_status, in_loantype, loanperiod, loan_type, cash_loan_volume, upgrade_loan_volume, existing_num_loans, allowed_ppl, decrease_rate)
    #here we save the parameters and result in a list for later processing
    pram_info.append((loan_status, in_loantype, loanperiod, loan_type, cash_loan_volume, upgrade_loan_volume, existing_num_loans, allowed_ppl, decrease_rate,result))
    return result

start_time = time.time()

result = gp_minimize(
    func=ob_lv,
    dimensions=param_space,
    n_calls=100,  # Number of optimization iterations
    random_state=60
)

print("Optimal Parameters:", result.x)
print("Optimal Value:", -result.fun) 
end_time = time.time()

elapsed_time_seconds = end_time - start_time
elapsed_time_minutes = elapsed_time_seconds / 60

print(f"The notebook has been running for {elapsed_time_seconds:.2f} seconds, or {elapsed_time_minutes:.2f} minutes.")

```
